Был выбран dataset с официального сайта coco https://cocodataset.org/#download

```python
import json
from pycocotools.coco import COCO

# Путь к файлу аннотаций
annotation_file = 'instances_train2014.json' 

coco = COCO(annotation_file)

# Получите список классов и количество объектов в каждом классе
cat_ids = coco.getCatIds()
cats = coco.loadCats(cat_ids)

for cat in cats:
    img_ids = coco.getImgIds(catIds=cat['id'])
    print('Класс: {}, количество объектов: {}'.format(cat['name'], len(img_ids)))
```

Визуализация

```python
import matplotlib.pyplot as plt
import cv2
import random

# случайное изображение
img = coco.loadImgs(coco.getImgIds()[random.randint(1,10)])[0]
img_folder = 'path_to_images/'
image = cv2.imread(img_folder + img['file_name'])
if image is None:
    print(f"Could not load image {img['file_name']}")
else:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    ann_ids = coco.getAnnIds(imgIds=img['id'])
    anns = coco.loadAnns(ann_ids)
    plt.imshow(image)
    plt.axis('off')
    coco.showAnns(anns)

    plt.show()
```
